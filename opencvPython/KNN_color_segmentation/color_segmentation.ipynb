```python
import cv2
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = 14, 7
```

```python
# it is gonna be a BGR colored image
img = cv2.imread("//home//salaheddineghamri//Downloads//Balls.jpg")
```

```python
reimg = cv2.resize(img, (600, 400))
```

```python
plt.imshow(reimg)
```

```python
# convert image to RGB colors
img = cv2.cvtColor(reimg, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
copyImage = img.copy()
```

```python
mylist = []
i = 0
for element in copyImage:
    mylist.append(element)
    if i == 100: break
    i+=1
```

```python
myimage = np.array(mylist)
```

```python
plt.imshow(myimage)
```

```python
cols = ['R', 'G', 'B']
Data=[]
shape = copyImage.shape
pxlnbr = shape[0]*shape[1]
for imgrow in copyImage:
    for pixel in imgrow:
        Data.append(pixel)
df = pd.DataFrame(columns=cols, data=Data, index=range(pxlnbr))
```

```python
kmeans = KMeans(n_clusters=6, random_state=0).fit(df)
```
